```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler

import pandas as pd

from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('income.csv')

df.head()
```

```python
plt.scatter(df['Age'],df['Income($)'])
```

```python
# use the k means method
km = KMeans(n_clusters=3)
```

```python
# directly fit and predict the data
y_pred = km.fit_predict(df[['Age','Income($)']])
```

```python
y_pred
```

```python
df['cluster'] = y_pred
```

```python
df
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

plt.scatter(df1.Age,df1['Income($)'],color='red')
plt.scatter(df2.Age,df2['Income($)'],color='black')
plt.scatter(df3.Age,df3['Income($)'],color='blue')

plt.legend()

# because of lack of proportion in scaling between x and y axis the cluster is not formed properly
```

```python
# DO the min-max scaling to reremove the clustering inaccuracies
scaler = MinMaxScaler()
scaler.fit(df[['Income($)']])

df['Income($)'] = scaler.transform(df[['Income($)']])

scaler.fit(df[['Age']])
df.Age = scaler.transform(df[['Age']])

df
```

```python
# TRAIN THE KMeans again to get data cluster

km = KMeans(n_clusters=3)

y_pred = km.fit_predict(df[['Age','Income($)']])
y_pred
```

```python
# calculate the clusters again and drop the previously calculated one

df['cluster'] = y_pred
df.drop('cluster',axis='columns')
df
```

```python
# check the clustering again

# IN REAL LIFE WE WILL HAVE TO USE ELBOW TECHNIQUE

df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

plt.scatter(df1.Age,df1['Income($)'],color='red')
plt.scatter(df2.Age,df2['Income($)'],color='black')
plt.scatter(df3.Age,df3['Income($)'],color='blue')

plt.xlabel('Age')
plt.ylabel('Income($)')

# to check the centroids of each cluster 
# get the clusters
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='orange',marker='*',label='centroid')

plt.legend()
```

```python
# TO PLOT THE ELBOW WE CALCULATE THE sse (sum of squared error) for n number of clusters
k_range = range(1,10)

sse = []

for k in k_range:
    km = KMeans(n_clusters=k)
#     exclude the name column
    km.fit(df[['Age','Income($)']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
# plot the SSE to check for the 'elbow'

plt.xlabel('K')
plt.xlabel('Sum of Squared Error')

plt.plot(k_range,sse)
# here k = 3 after checking the elbow of the plot
```
