```python
def gradient_descent(x, y):
    m_curr = b_curr = 0
    # decrease the number of iterations if the cost can be decreased further
    iterations = 10000
    n = 0
    # increase the learning_rate of iterations if the cost can be decreased further
    
    # NOTE: 0.08 derived after trial and error analysis
    #     learning_rate = 0.08
    
    learning_rate = 0.01

    if len(x) == len(y):
        n = len(x)
    else:
        return Exception('length of x and y dataset not equal')

#     plot the original data points x,y
    plt.scatter(x,y,color='red',marker='+',linewidth=5)

    
    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        
#         plot the iterations each time
        plt.plot(x,y_predicted,color='orange')

        # try to reduce cost at each iteration
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])

        # m derivative
        md = -(2 / n) * sum(x * (y - y_predicted))

        # b derivative
        bd = -(2 / n) * sum(y - y_predicted)

        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
    
    print(m_curr,b_curr)
        
```

```python
gradient_descent(x,y)
```
