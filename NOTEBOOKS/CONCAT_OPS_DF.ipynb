```python
import pandas as pd
```

```python
# df = pd.read_excel('sampledatafoodsales.xlsx','FoodSales')

bharat_weather = pd.DataFrame({
    "city":['Mumbai',"Indraprasth","Banglore"],
    "temperature":[32,45,30],
    "humidity":[80,60,78]
})
bharat_weather
```

```python
us_weather = pd.DataFrame({
    "city":['new york',"chicago","orlando"],
    "temperature":[21,14,35],
    "humidity":[68,65,75]
})
us_weather
```

```python
# ||ignore_index|| argument for getting continuous indexes
df = pd.concat([bharat_weather,us_weather],ignore_index=True)
df
```

```python
df = pd.concat([bharat_weather,us_weather],keys=['Bharat','US'])
df
```

```python
# get the subset/original dataframe back
df.loc["Bharat"]
```

```python
# ANOTHER USE CASE FOR CONCAT
```

```python
temp_df = pd.DataFrame({
    "city":['Mumbai','Indraprasth','Banglore'],
    "temperature":[42,33,25]
})
temp_df
```

```python
wind_df = pd.DataFrame({
    "city":['Mumbai','Indraprasth','Banglore'],
    "windspeed":[9,11,21]
})

wind_df
```

```python
# use axis argument to merge data and remove redundancy
df = pd.concat([temp_df,wind_df],axis=1)
df
```

```python
## ANOTHER USE CASE WHEN THE DATA IS NOT IN SAME ORDER IN COLUMNS
# use index to align data
# !! we can do index auomatically in merge
temp_df = pd.DataFrame({
    "city":['Indraprasth','Banglore','Mumbai'],
    "temperature":[42,33,25]
},index=[0,1,2])
temp_df

wind_df = pd.DataFrame({
    "city":['Mumbai','Indraprasth'],
    "windspeed":[9,11]
},index=[2,0])

wind_df

df = pd.concat([temp_df,wind_df],axis=1)
df
```

```python
event = pd.Series(['Humid','Dry','Rain'],name='event')
event
```

```python
df = pd.concat([temp_df,event],axis=1)
df
```
