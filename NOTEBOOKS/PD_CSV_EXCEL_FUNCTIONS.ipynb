```python
# FOR EXCEL now
```

```python
def convert_na_in_cells(cell):
    if cell == 'n.a':
        return None
    return cell

def convert_dateTime_to_date(date):
    return pd.to_datetime(date).date()
```

```python
# using functions to filter data in excel files
df2 = pd.read_excel('sampledatafoodsales.xlsx','FoodSales',converters ={
    'Quantity':convert_na_in_cells,
    'OrderDate':convert_dateTime_to_date
}).head(30)
```

```python
df2
```

```python
df2.to_excel('demo_samplefooddatasales.xlsx',index=False,sheet_name="FoodSales_Demo",startrow=0,startcol=0)
```
