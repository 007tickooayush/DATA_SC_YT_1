```python
# AXES, LABELS, LEGENDS, GRIDS

days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [32,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.xlabel("Days")
plt.ylabel("Temps")
plt.title('Weather')

plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")

# best = default value
plt.legend(loc="best",shadow=True,fontsize='small') # to show the labels defined for each plot
# GRID
plt.grid()
```

```python
import numpy as np

# BAR CHART
company = ['GOOGL','AMZN','MSFT','FB']
revenue = [90,137,89,27]
profit= [40,2,34,12]

plt.ylabel("revenue(bln)")
plt.title("US TECH STOCKS")

# setting up to get String labels in the bar chart
xpos = np.arange(len(company))
plt.xticks(xpos,company)

plt.bar(xpos-0.2,revenue,width=0.4,label='Revenue')
plt.bar(xpos+0.2,profit,width=0.4,label='Profit')

plt.legend()
```

```python
# HORIZONTAL BAR CHART

plt.title("US TECH STOCKS")

ypos = np.arange(len(company))
plt.yticks(ypos,company)

plt.barh(xpos-0.2,revenue,label='Revenue')
plt.barh(xpos+0.2,profit,label='Profit')

plt.legend()
```

```python
# HISTOGRAMS

blood_sugar_men = [113,85,90,150,150,149,89,93,115,135,80,77,82]
blood_sugar_women = [67,98,89,120,133,150,84,69,89,79,120,112,100]

# default usage =>
# plt.hist(blood_sugar_mem)

# plt.hist(blood_sugar_men,bins=[80,100,125,150],rwidth=0.95,color='orange',histtype='step')

plt.xlabel('Sugar Range')
plt.ylabel('Total Patients')
plt.title('Blood Sugar Levels Data')

plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['red','orange'],label=['men','women']
         ,orientation='horizontal')
plt.legend()
```

```python
# PIE CHARTS

exp_vals = [1400,600,300,410,250]
exp_labels = ['Home Rent','Food','Phone/Internet bill','Car','Other Utilities']

plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.2f%%',explode=[0,0.1,0.1,0.2,0.5],startangle=180)
plt.show() # to remove the descriptive text
```
